```python
import cloudscraper
from bs4 import BeautifulSoup as bs 

# Scrape data for first page
url = 'https://batdongsan.com.vn/nha-dat-ban-tp-hcm'

scraper = cloudscraper.create_scraper(delay=10, browser="chrome") 
content = scraper.get(url)
print(content.status_code)
print(content.text)

content = content.text
```

```python
# Get number of page
soup = bs(content, 'html.parser')
num_of_page = soup.find_all('a', class_='re__pagination-number')

# Get url list in first page
url_list = []
url_list = soup.find_all('a', class_='js__product-link-for-product-id')
```

```python
print(num_of_page)
print(url_list)
```
