To *ask meaningful question*, I have to know what I was interested in. That was a hard thing to do for me, because I tend to ask very broad questions and not quite knowing what I wanted to to **know**. After working on the project, there are some tips I've developed to help:

> From what character's point of view am I viewing the data?

Essentially what this is saying is I'm treating a dataset like a story, and I have to choose a character in that story's perspective to make sense of anything. There have to be a constant (perspective) to make sense of things because people understand things in the form of story. In my analysis, I take a perspective of a young real estate broker who want to understand the real estate renting market in Ho Chi Minh city. I found defining the perspective is very important, becuase asking a question from a my point of view (a person who barely know anything about real estate) is different from the perspective of my parents (home owners) or of real estate agents (have access to large and accurate real estate data).

In short, it helps me narrow down on what questions **I want to ask** and what answer **I can understand**. 



> Chop the question into smaller pieces but make it clear what you want to answer

In my analysis, you can see that I ask a big question and break it downs to smaller questions to understand it better. What you don't know is when I did that at first, I still didn't know what I wanted to ask. Only after I could provide an example of very specific questions under that question can I understand what answer. For example, take the first sub question of the first question that I post.

<span style="color:grey">*How are the number of real estate for rent distributed across districts in Ho Chi Minh city?*</span>

This is a very vague and over the place question. We don't know what is the subject and what is the action. To do that, I have to define the question further.

<span style="color:grey">
In other word, we are asking: </br>
- Which districts has the most real estate property for rent? </br>
- Which districts has the least real estate property for rent?  
</span>

Defining the question further help us identify that the subject of the question is the **district**, and the action if to **rank/ classify** the district in a certain order.

And finally concluding in a questions that actually lead us to the answer for the initial question that we post.

<span style="color:grey">
=> <b>How big</b> is <b>the renting market</b> in <b>each districts</b> and how are they compared to each other.
</span>

This process help me understand the shape of the question and the data better, but I'm still not very good at it. Reading others analysis has been helful with this, and I believe this is a skills to build over time.
